```python
import requests
```

## Список объектов

```python
offset_ = 1
all_data = []
limit_ = 1000
for a in range(10):
    url = f'https://xn--80az8a.xn--d1aqf.xn--p1ai/%D1%81%D0%B5%D1%80%D0%B2%D0%B8%D1%81%D1%8B/api/kn/object?offset={offset_}&limit={limit_}&sortField=devId.devShortCleanNm&sortType=asc&objStatus=0'
    resp = requests.get(url, timeout=300)
    if resp.status_code == 200:
        objects_data = resp.json().get('data').get('list')
        all_data += objects_data
        offset_ += 1000
    else:
        print(resp.status_code)
print(len(all_data))
```

```python
objids = [x.get('objId') for x in all_data]
```

```python
# резервная копия в CSV

import csv, json
with open('res.csv', 'a') as file:
    wr = csv.writer(file)
    for line in objects_full:
        wr.writerow([json.dumps(line)])
```

```python
objects_full = []
import csv, json
with open('res.csv', encoding="utf-8") as file:
    rd = csv.reader(file)
    for row in rd:
        objects_full.append(json.loads(row[0]))
print(objects_full[0])
```

```python
import pandas as pd
df = pd.json_normalize(objects_full)
```
